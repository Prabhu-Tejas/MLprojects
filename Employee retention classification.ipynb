```python
X= data.drop(['left'],axis=1)
X
```

```python
X.info()
```

```python
X.rename(columns={'sales':'Department'}, inplace=True)
X
```

```python
A=X[["number_project","Work_accident","promotion_last_5years","Department", "salary"]].astype('category')
A
```

```python
X.drop(columns=["number_project","Work_accident","promotion_last_5years","Department", "salary"], axis=1, inplace= True)
X
#Standardising to keep data on common scale
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(X)
X_ss=pd.DataFrame(ss.transform(X), columns=['satisfaction_level','last_evaluation','average_montly_hours','time_spend_company'])
X_ss
```

```python
X=pd.concat([A,X_ss], axis=1)
X
```

One Hot encoding on categorical data

```python
X=pd.get_dummies(X)
X
```

```python
Y
```

Spliting the data

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest= train_test_split(X,Y,test_size=0.15)
```

Model building

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(Xtrain,Ytrain)
```

```python
ypred=log_reg.predict(Xtest)
print(ypred)
```

```python
Ytest
```

Model Performance Analysis

```python
from sklearn.metrics import confusion_matrix
ypred= log_reg.predict(Xtest)
confusion_matrix(Ytest,ypred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Ytest,ypred))
```
